```python
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
from dotenv import load_dotenv
load_dotenv()
# 타빌리 서치 도구를 바인딩한 모델 사용하기

# 1. 도구 생성
tool = TavilySearch(max_result=2)
tools = [tool]

# 2. llm 객체 생성 후 객체에 도구를 바인딩
llm = ChatOpenAI(model='gpt-4o-mini')
llm_with_tools = llm.bind_tools(tools)
```

```python
# 1) 도구 사용이 필요 없는 질문 -> 웹검색이 필요없는 질문
llm_with_tools.invoke('안녕')

# tool_calls=[]라는 결과값 -> 도구 호출을 안함. 웹 서치를 안했다는 의미
```

```python
# 2) 도구 사용이 필요한 질문 -> 웹 검색이 필요한 질문
llm_with_tools.invoke('2026년 AI 트랜드는 무엇인가요?')
```

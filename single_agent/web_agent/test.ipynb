```python
from langchain.tools import tool

# 랭체인에서 도구를 구현하기 위해서는 tool 데코레이터(@tool)를 통해 함수로 사용할 도구를 직접 정의
# 함수 선언 아랫줄에 독스트링 docstring을 통해 간단히 도구의 설명을 작성. 제작자가 아니라 llm이 읽을 것임.
# 이 설명 텍스트는 실제로 도구 호출에 사용되므로 꼼꼼히 작성해두는 것이 좋음. 

@tool
def add(a: int, b: int) -> int:
    """Adds a and b.

    Args:
        a: first int
        b: second int
    """ 
    
    return a + b

@tool
def multiply(a:int, b:int)-> int:
    """Multiplies a and b.

    Args:
        a: first int
        b: second int
    """
    return a * b

tools = [add, multiply]
```

```python
# LLM 에게 사용할 도구 쥐어주기
from dotenv import load_dotenv
load_dotenv()

# bind_tools는 랭체인 채팅 모델이 구현하는 메서드로, 도구가 필요한 경우 도구 호출을 하게 함.
# 즉, 도구와 모델을 연결하는 인터페이스로 LLM 에게 사용할 도구를 결합해 주는 용도.
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-4o-mini')
llm_with_tools = llm.bind_tools(tools) # llm에게 사용할 수 있는 도구를 전달
```

```python
response = llm.invoke('대한민국의 수도는?')
response
```

```python
query = "3 곱하기 5는 뭔가요? 그리고 2 더하기 4는 뭔가요?"

response = llm_with_tools.invoke(query)
response
```
